```python
# KNN imputation
from sklearn.impute import KNNImputer

imp = KNNImputer(n_neighbors=2, weights='uniform')
cols_to_impute = []
for x in x_features:
    
    if train_df[x].isnull().any():
        cols_to_impute.append(x)
print(cols_to_impute)
train_df[cols_to_impute] = imp.fit_transform(train_df[cols_to_impute])
```

```python
# test dataset
cols_to_impute = []
for x in x_features:

    if test_df[x].isnull().any():
        if test_df[x].notnull().any():
            cols_to_impute.append(x)
            
        else:
            test_df[x] = 0
print(cols_to_impute)
test_df[cols_to_impute] = imp.fit_transform(test_df[cols_to_impute])
```

```python
# Scale numeric values
from sklearn.preprocessing import StandardScaler

train_df_transformed = train_df.copy()
test_df_transformed = test_df.copy()
x_features = train_df.drop(['State_FIPS_Code', 'County_FIPS_Code', 'CHSI_County_Name', 'CHSI_State_Abbr', 'Diabetes'], axis=1).columns.tolist()
numeric_cols = [x for x in x_features if 'Ind' not in x]
numeric_cols.append('Diabetes')

scaler = StandardScaler()
train_df_transformed[numeric_cols] = scaler.fit_transform(train_df[numeric_cols])
test_df_transformed[numeric_cols] = scaler.fit_transform(test_df[numeric_cols])
```

### Scoring Bins
 
0 - Std. Score < -0.81: **Low Risk**

1 - Std. Score between -.81 and 1.03 **Medium Risk**

2 - Std. Score between 1.03 and 2.86 **High Risk**

3 - Std. Score > 2.86 **Urgent Risk**

```python
def label_score(rate):
    if rate < -0.81:
        return 0
    elif rate > -0.81 and rate < 1.03:
        return 1
    elif rate > 1.03 and rate < 2.86:
        return 2
    elif rate > 2.86:
        return 3
```

```python
train_df_transformed['Diabetes_Risk'] = train_df_transformed['Diabetes'].apply(lambda x: label_score(x))
test_df_transformed['Diabetes_Risk'] = test_df_transformed['Diabetes'].apply(lambda x: label_score(x))
```

```python
from sklearn.model_selection import train_test_split
x_features = train_df_transformed.drop(['State_FIPS_Code', 'County_FIPS_Code', 'CHSI_County_Name', 'CHSI_State_Abbr', 'Diabetes_Risk'], axis=1).columns.tolist()
pred_var = 'Diabetes_Risk'

# 0.7 train test split
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(train_df_transformed[x_features], train_df_transformed[pred_var], test_size=0.7, random_state=42)
```

```python
from sklearn import metrics
import seaborn as sns

def evaluate_multi_classification_model(ytest, ypred, yscore):
    """Views variety of model evaluation methods, including accuracy, recall, precision, confusion matrix and RMSE"""
    acc = metrics.accuracy_score(ytest, ypred)
    prec = metrics.precision_score(ytest, ypred, average='weighted')
    rec = metrics.recall_score(ytest, ypred, average='weighted')
    f1 = metrics.f1_score(ytest, ypred, average='weighted')
    conf_matrix = metrics.confusion_matrix(ytest, ypred)

    evaluation_df = pd.DataFrame({'Metric': [ 'Accuracy', 'Precision', 'Recall', 'F1'],
                                 'Score': [acc, prec, rec, f1]})


    #ROC and AUC
    fpr = dict()
    tpr = dict()
    roc_auc = dict()

    ytest_dummies = pd.get_dummies(ytest, drop_first = False).values

    for i in list(set(ytest)):
        fpr[i], tpr[i], thresholds = metrics.roc_curve(ytest_dummies[:,i], yscore[:,i])
        roc_auc[i] = metrics.auc(fpr[i], tpr[i])

    return evaluation_df, conf_matrix, fpr, tpr, roc_auc

def plot_conf_matrix(conf_matrix):
    plt.figure(figsize=(14,8))
    sns.heatmap(conf_matrix, square=True, annot=True,
                annot_kws={'size':15}, cmap='YlGnBu', cbar=False)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
    plt.clf()
    
def plot_multi_auc_roc(model_type, fpr, tpr, roc_auc):
    fig, ax = plt.subplots(figsize=(14,8))
    ax.plot([0,1],[0,1],'r--')
    for i in list(fpr.keys()):
        ax.plot(fpr[i], tpr[i], label='ROC %0.2f for label %i'%(roc_auc[i], i))
    ax.set_title(model_type)
    ax.legend(loc='lower right')
    ax.set_xlim([-0.1,1.2])
    ax.set_ylim([-0.1,1.2])
    ax.set_ylabel('True Positive Rate')
    ax.set_xlabel('False Positive Rate')
    plt.show()
```

### Logistic Regression

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=10000)
lr.fit(xtrain, ytrain)
ypred = lr.predict(xtest)
yscore = lr.predict_proba(xtest)
```

```python
eval_df, conf_matrix, fpr, tpr, roc_auc = evaluate_multi_classification_model(ytest, ypred, yscore)
print(eval_df)
plot_conf_matrix(conf_matrix)
plot_multi_auc_roc('Logistic Regression', fpr, tpr, roc_auc)
```
