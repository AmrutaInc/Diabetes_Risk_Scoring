```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 1000)
```

```python
train_df = pd.read_csv('diabetes_train.csv').drop('Unnamed: 0', axis=1)
test_df = pd.read_csv('diabetes_test.csv').drop('Unnamed: 0', axis=1)
```

```python
#replace negative values with NA
def replace_with_na(col):
    result = [x if x > 0 else None for x in col]
    return result

x_features = train_df.drop(['State_FIPS_Code', 'County_FIPS_Code', 'CHSI_County_Name', 'CHSI_State_Abbr', 'Diabetes'], axis=1).columns.tolist()

train_df[x_features] = train_df[x_features].apply(replace_with_na)
test_df[x_features] = test_df[x_features].apply(replace_with_na)
```

```python
# KNN imputation
from sklearn.impute import KNNImputer

imp = KNNImputer(n_neighbors=2, weights='uniform')
cols_to_impute = []
for x in x_features:
    
    if train_df[x].isnull().any():
        cols_to_impute.append(x)
print(cols_to_impute)
train_df[cols_to_impute] = imp.fit_transform(train_df[cols_to_impute])
```

```python
# test dataset
cols_to_impute = []
for x in x_features:

    if test_df[x].isnull().any():
        if test_df[x].notnull().any():
            cols_to_impute.append(x)
            
        else:
            test_df[x] = 0
print(cols_to_impute)
test_df[cols_to_impute] = imp.fit_transform(test_df[cols_to_impute])
```

```python
# Scale numeric values
from sklearn.preprocessing import StandardScaler

train_df_transformed = train_df.copy()
test_df_transformed = test_df.copy()
x_features = train_df.drop(['State_FIPS_Code', 'County_FIPS_Code', 'CHSI_County_Name', 'CHSI_State_Abbr', 'Diabetes'], axis=1).columns.tolist()
numeric_cols = [x for x in x_features if 'Ind' not in x]
numeric_cols.append('Diabetes')

scaler = StandardScaler()
train_df_transformed[numeric_cols] = scaler.fit_transform(train_df[numeric_cols])
test_df_transformed[numeric_cols] = scaler.fit_transform(test_df[numeric_cols])
```

```python
# bin diabetes
import matplotlib.pyplot as plt
n, bins, patches = plt.hist(train_df_transformed.Diabetes, bins=5)
plt.title('Standardized Diabetes Rate Histogram')
plt.xlabel('Standardized Diabetes Rate')
plt.ylabel('Frequency')
```

## Scoring Bins

2 classes: non-risk vs at-risk

0. standardized diabetes rate is <= 0
1. standardized diabetes rate is > 0

```python
def label_score(rate):
    if rate <= 0:
        return 0
    elif rate > 0:
        return 1
```

```python
train_df_transformed['Diabetes_Risk'] = train_df_transformed['Diabetes'].apply(lambda x: label_score(x))
test_df_transformed['Diabetes_Risk'] = test_df_transformed['Diabetes'].apply(lambda x: label_score(x))
```

```python
train_df_transformed.drop('Diabetes', axis=1, inplace=True)
```

```python
train_df_transformed['Diabetes_Risk'].value_counts()
```

```python
test_df_transformed.to_csv('diabetes_BI_RISK_val.csv')
```

```python
from sklearn.model_selection import train_test_split
x_features = train_df_transformed.drop(['State_FIPS_Code', 'County_FIPS_Code', 'CHSI_County_Name', 'CHSI_State_Abbr', 'Diabetes_Risk'], axis=1).columns.tolist()
pred_var = 'Diabetes_Risk'

# 0.7 train test split
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(train_df_transformed[x_features], train_df_transformed[pred_var], test_size=0.7, random_state=42)
```

```python
from sklearn import metrics
import seaborn as sns

def evaluate_multi_classification_model(ytest, ypred, yscore):
    """Views variety of model evaluation methods, including accuracy, recall, precision, confusion matrix and RMSE"""
    acc = metrics.accuracy_score(ytest, ypred)
    prec = metrics.precision_score(ytest, ypred, average='weighted')
    rec = metrics.recall_score(ytest, ypred, average='weighted')
    f1 = metrics.f1_score(ytest, ypred, average='weighted')
    conf_matrix = metrics.confusion_matrix(ytest, ypred)

    evaluation_df = pd.DataFrame({'Metric': [ 'Accuracy', 'Precision', 'Recall', 'F1'],
                                 'Score': [acc, prec, rec, f1]})


    #ROC and AUC
    fpr = dict()
    tpr = dict()
    roc_auc = dict()

    ytest_dummies = pd.get_dummies(ytest, drop_first = False).values

    for i in list(set(ytest)):
        fpr[i], tpr[i], thresholds = metrics.roc_curve(ytest_dummies[:,i], yscore[:,i])
        roc_auc[i] = metrics.auc(fpr[i], tpr[i])

    return evaluation_df, conf_matrix, fpr, tpr, roc_auc

def plot_conf_matrix(conf_matrix):
    plt.figure(figsize=(14,8))
    sns.heatmap(conf_matrix, annot=True, fmt="d",
                annot_kws={'size':15}, cmap='YlGnBu')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
    plt.clf()
    
def plot_multi_auc_roc(model_type, fpr, tpr, roc_auc):
    fig, ax = plt.subplots(figsize=(14,8))
    ax.plot([0,1],[0,1],'r--')
    for i in list(fpr.keys()):
        ax.plot(fpr[i], tpr[i], label='ROC %0.2f for label %i'%(roc_auc[i], i))
    ax.set_title(model_type)
    ax.legend(loc='lower right')
    ax.set_xlim([-0.1,1.2])
    ax.set_ylim([-0.1,1.2])
    ax.set_ylabel('True Positive Rate')
    ax.set_xlabel('False Positive Rate')
    plt.show()
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=10000)
lr.fit(xtrain, ytrain)
ypred = lr.predict(xtest)
yscore = lr.predict_proba(xtest)
```

```python
eval_df, conf_matrix, fpr, tpr, roc_auc = evaluate_multi_classification_model(ytest, ypred, yscore)
print(eval_df)
plot_conf_matrix(conf_matrix)
plot_multi_auc_roc('Logistic Regression', fpr, tpr, roc_auc)
```

```python
# Light GBM
import lightgbm as lgb

lfb_clf = lgb.LGBMClassifier()
lfb_clf.fit(xtrain, ytrain)
ypred = lfb_clf.predict(xtest)
yscore = lfb_clf.predict_proba(xtest)
```

```python

lgb.plot_importance(lfb_clf, max_num_features=20, height=0.5, figsize=(12,8))
plt.show()
```

```python
eval_df, conf_matrix, fpr, tpr, roc_auc = evaluate_multi_classification_model(ytest, ypred, yscore)

print(eval_df)

plot_conf_matrix(conf_matrix)

plot_multi_auc_roc('Light GBM', fpr, tpr, roc_auc)
```

```python
# Cross Validation

def cv_lgb(xtrain, ytrain, xtest, ytest):
    # Parameters
    N_FOLDS = 3
    MAX_BOOST_ROUNDS = 700
    LEARNING_RATE = .0022

    params = {
        "objective" : "multiclass",
        'num_class':4,
        "metric" : "multi_error",
        "num_leaves" : 64,
        "learning_rate" : 0.005,
        "bagging_fraction" : 0.7,
        "feature_fraction" : 0.5,
        "bagging_frequency" : 6,
        "bagging_seed" : 42,
        "verbosity" : 1,
        "seed": 42,
            }

    dtrain = lgb.Dataset(xtrain, label=ytrain)
    dtest = lgb.Dataset(xtest, label=ytest)

    evals_result = {}
    model_opt = lgb.train(params, dtrain, num_boost_round=MAX_BOOST_ROUNDS, valid_sets=dtest,
                       verbose_eval=20, early_stopping_rounds=100, evals_result = evals_result)

    print('Current parameters:\n', params)

    return model_opt, params
```

```python
model_opt, params = cv_lgb(xtrain, ytrain, xtest, ytest)
```

```python
yscore = model_opt.predict(xtest, num_iteration=model_opt.best_iteration)
ypred = np.array(pd.DataFrame(yscore).eq(pd.DataFrame(yscore).max(axis=1), 0).dot(pd.DataFrame(yscore).columns))

eval_df, conf_matrix, fpr, tpr, roc_auc = evaluate_multi_classification_model(ytest, ypred, yscore)

print(eval_df)

plot_conf_matrix(conf_matrix)

plot_multi_auc_roc('Light GBM Optimized',fpr, tpr, roc_auc)
```

### Conclusion

We see that we get pretty solid results when looking at the problem as a binary classification problem. Our best accuracy with Light GBM is 70.9%, with an F1 score of 70.8%. We get an ROC of .78 for both non-risk and at-risk labels.
